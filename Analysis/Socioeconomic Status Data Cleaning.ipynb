```python
import pandas as pd
import numpy as np
```

```python
raw_socioeconomic_data_df = pd.read_csv("./data/CA_modified_data_for_socioeconomic_data.csv")
raw_socioeconomic_data_df
```

```python
california_data = raw_socioeconomic_data_df.loc[((raw_socioeconomic_data_df["LocationDesc"] == "California") 
                                                 & (raw_socioeconomic_data_df["Class"] == "Obesity / Weight Status"))]
california_data = california_data.rename({"LocationAbbr":"State (Abbreviation)",
                        "LocationDesc":"State",
                        "Datasource":"Source",
                        "Question":"Have Obesity/Overweight Classification",
                        "Data_Value":"Obesity Percentage",
                        "Sample_Size":"Number of Samples",
                        "Education":"Level of Education"
                       }, axis='columns')

california_data.loc[("have obesity" in string for string in california_data["Have Obesity/Overweight Classification"].astype("string")), "Have Obesity/Overweight Classification"] = "obesity"

california_data
```

```python
california_data = california_data[["YearStart", "YearEnd", "State", "Class","Have Obesity/Overweight Classification", "Obesity Percentage", "Number of Samples", "Age(years)", "Level of Education", "Gender", "Income", "Race/Ethnicity"]]
california_data
```

```python
age_ranges = california_data["Age(years)"].unique()
print(age_ranges)
education_levels = california_data["Level of Education"].unique()
print(education_levels)
gender = california_data["Gender"].unique()
print(gender)
income_ranges = california_data["Income"].unique()
print(income_ranges)
race_ethnicity = california_data["Race/Ethnicity"].unique()
print(race_ethnicity)
```

```python
age_data = california_data.dropna(subset = ["Age(years)"])
age_data = age_data[["YearEnd", "State", "Class", "Have Obesity/Overweight Classification", "Obesity Percentage", "Number of Samples", "Age(years)"]]
age_data.head(20)
```

```python
level_of_education_data = california_data.dropna(subset = ["Level of Education"])
level_of_education_data = level_of_education_data[["YearEnd", "State", "Class","Have Obesity/Overweight Classification", "Obesity Percentage", "Number of Samples", "Level of Education"]]
level_of_education_data.head(20)
```

```python
gender_data = california_data.dropna(subset = ["Gender"])
gender_data = gender_data[["YearEnd", "State", "Class","Have Obesity/Overweight Classification", "Obesity Percentage", "Number of Samples", "Gender"]]
gender_data.head(20)
```

```python
income_data = california_data.dropna(subset = ["Income"])
income_data = income_data[["YearEnd", "State", "Class","Have Obesity/Overweight Classification", "Obesity Percentage", "Number of Samples", "Income"]]
income_data = income_data.loc[income_data["Income"] != "Data not reported"]
income_data.head(20)
```

```python
race_ethnicity_data = california_data.dropna(subset = ["Race/Ethnicity"])
race_ethnicity_data = race_ethnicity_data[["YearEnd", "State", "Class","Have Obesity/Overweight Classification", "Obesity Percentage", "Number of Samples", "Race/Ethnicity"]]
race_ethnicity_data = race_ethnicity_data.dropna()
race_ethnicity_data.head(20)
```
