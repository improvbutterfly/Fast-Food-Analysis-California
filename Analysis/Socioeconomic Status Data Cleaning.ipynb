```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt



```

```python
raw_socioeconomic_data_df = pd.read_csv("./data/CA_modified_data_for_socioeconomic_data.csv")
raw_socioeconomic_data_df
```

```python
california_data = raw_socioeconomic_data_df.loc[((raw_socioeconomic_data_df["LocationDesc"] == "California") 
                                                 & (raw_socioeconomic_data_df["Class"] == "Obesity / Weight Status"))]
california_data = california_data.rename({"LocationAbbr":"State (Abbreviation)",
                        "LocationDesc":"State",
                        "Datasource":"Source",
                        "Question":"Classification",
                        "Data_Value":"Obesity Percentage",
                        "Sample_Size":"Number of Samples",
                        "Education":"Level of Education"
                       }, axis='columns')

california_data.loc[("have obesity" in string for string in california_data["Classification"].astype("string")), "Have Obesity/Overweight Classification"] = "obesity"

california_data
```

```python
california_data = california_data[["YearStart", "YearEnd", "State", "Class","Classification", "Obesity Percentage", "Number of Samples", "Age(years)", "Level of Education", "Gender", "Income", "Race/Ethnicity"]]
california_data
```

```python
age_ranges = california_data["Age(years)"].unique()
print(age_ranges)
education_levels = california_data["Level of Education"].unique()
print(education_levels)
gender = california_data["Gender"].unique()
print(gender)
income_ranges = california_data["Income"].unique()
print(income_ranges)
race_ethnicity = california_data["Race/Ethnicity"].unique()
print(race_ethnicity)
```

```python
age_data = california_data.dropna(subset = ["Age(years)"])
age_data = age_data[["YearEnd", "State", "Class", "Classification", "Obesity Percentage", "Number of Samples", "Age(years)"]]
age_data.head(20)
```

```python
level_of_education_data = california_data.dropna(subset = ["Level of Education"])
level_of_education_data = level_of_education_data[["YearEnd", "State", "Class","Classification", "Obesity Percentage", "Number of Samples", "Level of Education"]]
level_of_education_data.head(20)
```

```python
gender_data = california_data.dropna(subset = ["Gender"])
gender_data = gender_data[["YearEnd", "State", "Class","Classification", "Obesity Percentage", "Number of Samples", "Gender"]]
gender_data.head(20)
```

```python
income_data = california_data.dropna(subset = ["Income"])
income_data = income_data[["YearEnd", "State", "Class","Classification", "Obesity Percentage", "Number of Samples", "Income"]]
income_data = income_data.loc[income_data["Income"] != "Data not reported"]
income_data.head(20)
```

```python
race_ethnicity_data = california_data.dropna(subset = ["Race/Ethnicity"])
race_ethnicity_data = race_ethnicity_data[["YearEnd", "State", "Class","Classification", "Obesity Percentage", "Number of Samples", "Race/Ethnicity"]]
race_ethnicity_data = race_ethnicity_data.dropna()
race_ethnicity_data.head(20)
```

```python
#we have data by age, education level, gender, income, and race/ethnicity. we need to generate graphs for all 
#these data points vs percent obesity, for years 2011 & 2016. Additionally, we are only concerned for those that
#fall into the "obese" category, not overweight

#we need to use pd.loc to filter data in years 2011 and 2016, as well as to get the data for only obese

#age data
age_data2011 = age_data.loc[((age_data["Classification"] == "obesity") & (age_data["YearEnd"] == 2011))]
age_data2016 = age_data.loc[((age_data["Classification"] == "obesity") & (age_data["YearEnd"] == 2016))]
#level of education
level_of_education_data2011 = level_of_education_data.loc[((level_of_education_data["Classification"] == "obesity") & (level_of_education_data["YearEnd"] == 2011))]
level_of_education_data2016 = level_of_education_data.loc[((level_of_education_data["Classification"] == "obesity") & (level_of_education_data["YearEnd"] == 2016))]
#gender data
gender_data2011 = gender_data.loc[((gender_data["Classification"] == "obesity") & (gender_data["YearEnd"] == 2011))]
gender_data2016 = gender_data.loc[((gender_data["Classification"] == "obesity") & (gender_data["YearEnd"] == 2016))]
#income data
income_data2011 = income_data.loc[((income_data["Classification"] == "obesity") & (income_data["YearEnd"] == 2011))]
income_data2016 = income_data.loc[((income_data["Classification"] == "obesity") & (income_data["YearEnd"] == 2016))]
#race data
race_ethnicity_data2011 = race_ethnicity_data.loc[((race_ethnicity_data["Classification"] == "obesity") & (race_ethnicity_data["YearEnd"] == 2011))]
race_ethnicity_data2016 = race_ethnicity_data.loc[((race_ethnicity_data["Classification"] == "obesity") & (race_ethnicity_data["YearEnd"] == 2016))]
```

```python
age_data2011 = age_data.reindex([2708, 2709, 2710, 2711, 2568])
age_data2011
```

```python
#now, we need to graph the data for each, starting with 2011 data

#AGE VS OBESITY
age_range_2011 = age_data2011["Age(years)"]
values_by_age2011 = age_data2011["Obesity Percentage"]
plt.bar(age_range_2011, values_by_age2011)
plt.xlabel('Age(years)')
plt.ylabel('Obesity Percentage')
plt.title("Percentage of Obese people by Age in 2011")
plt.savefig("./output images/age_vs_obesity_2011.png")
plt.show()

# age_data2011.head()
```

```python
#EDUCATION VS OBESITY
education_range_2011 = level_of_education_data2011["Level of Education"]
values_by_education_range2011 = level_of_education_data2011["Obesity Percentage"]
plt.bar(education_range_2011, values_by_education_range2011)
plt.xticks(education_range_2011, rotation = 15)
plt.xlabel("Level of Education")
plt.ylabel('Obesity Percentage')
plt.title("Percentage of Obese people by Level of Education in 2011")
plt.savefig("./output images/education_vs_obesity_2011.png")
plt.show()
```

```python
#GENDER VS OBESITY
gender_range_2011 = gender_data2011["Gender"]
values_by_gender2011 = gender_data2011["Obesity Percentage"]
plt.bar(gender_range_2011, values_by_gender2011)
plt.xlabel('Gender')
plt.ylabel('Obesity Percentage')
plt.title("Percentage of Obese people by Gender in 2011")
plt.savefig("./output images/gender_vs_obesity_2011.png")
plt.show()
```

```python
#INCOME VS OBESITY
income_range_2011 = income_data2011["Income"]
values_by_income2011 = income_data2011["Obesity Percentage"]
plt.bar(income_range_2011, values_by_income2011)
plt.xticks(income_range_2011, rotation = 45)
plt.xlabel('Income')
plt.ylabel('Obesity Percentage')
plt.title("Percentage of Obese people by Income in 2011")
plt.savefig("./output images/Income_vs_obesity_2011.png")
plt.show()
```

```python
#RACE VS OBESITY
race_ethnicity_2011 = race_ethnicity_data2011["Race/Ethnicity"]
values_by_race2011 = race_ethnicity_data2011["Obesity Percentage"]
plt.bar(race_ethnicity_2011, values_by_race2011)
plt.xticks(race_ethnicity_2011, rotation = 70)
plt.xlabel('Race/Ethnicity')
plt.ylabel('Obesity Percentage')
plt.title("Percentage of Obese people by Race/Ethnicity in 2011")
plt.savefig("./output images/race_vs_obesity_2011.png")
plt.show()
```
