```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as palt
import requests
from census import Census
# Census API Key
import configparser
config = configparser.ConfigParser()
config.read('config.py')
api_key = config['DEFAULT']['API_KEY']

from census import Census
c = Census(api_key, year=2016)
```

```python
# Run Census Search to retrieve data on all zip codes (2013 ACS5 Census)
# See: https://github.com/CommerceDataService/census-wrapper for library documentation
# See: https://gist.github.com/afhaque/60558290d6efd892351c4b64e5c01e9b for labels
#use the Census wrapper to get information from the Census API
census_data = c.acs5.get(("NAME", "B01003_001E","B15003_002E","B15003_017E","B15003_018E","B15003_021E","B15003_022E","B15003_023E","B15003_024E","B15003_025E"), {'for': 'county:*'})

#set the Census data from the wrapper into a Pandas DataFrame
census_pd = pd.DataFrame(census_data)

print(len(census_pd))
census_pd.head()
```

*how to split the column into county and into state<br>
*then you can filter out by state

```python
#rename respective columns into data
census_pd = census_pd.rename(columns={"B15003_002E": "No Education", 
                                      "B01003_001E": "Population",
                                        "B15003_017E": "High School Education",
                                        "B15003_018E": "GED education",
                                        "B15003_021E": "Associate Education",
                                        "B15003_022E": "Bachelor Education",
                                        "B15003_023E": "Masters Education",
                                        "B15003_024E": "Professional Education",
                                        "B15003_025E": "Doctorate Education",
                                        "NAME": "Name"}) 
```

```python
census_pd.head()
```

```python
#filter out only data for California
california_data = census_pd.loc[census_pd['state'] == "06"]
california_data.head()
```

```python
california_data.head()
```

```python
#split the county to remove the word County from the county section
for index, row in california_data.iterrows():
    county = row["County"]
    county_split = county.split(" County")
    california_data.loc[index, "County"] = str(county_split[0])
        
california_data.head()
```

```python
del california_data["state"]
del california_data["county"]
```

```python
california_data = california_data.sort_values(by=['County'])
california_data.head()
```

```python
#get a new column for number of people with GED, Associate Education, Bachelor, Masters, Professional and Doctorate
california_data['College'] = california_data['Associate Education'] + california_data['Bachelor Education'] + california_data['Masters Education'] + california_data['Professional Education'] + california_data['Doctorate Education']
```

```python
california_data = california_data.reset_index(drop=True)
california_data.head()
```

```python
#from the new column of people with education degrees to get the rate of the population within a county
california_data["College Rate"] = 100 * \
   california_data["College"].astype(
       int) / california_data["Population"].astype(int)
```

```python
california_data.head()
```

```python
# Save as a csv
# Note to avoid any issues later, use encoding="utf-8"
california_data.to_csv("census_education_data_2016.csv", encoding="utf-8", index=False)
```
