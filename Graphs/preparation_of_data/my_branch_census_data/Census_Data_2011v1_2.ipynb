*how to split the column into county and into state<br>
*then you can filter out by state

```python
california_data = census_pd.loc[census_pd['state'] == "06"]
california_data
```

```python
# drop state and county code columns
california_data = california_data.drop(columns={"state", "county"})
california_data
```

```python
# Remove "County, California" from County column
for index, row in california_data.iterrows():
    county = row["County"]
    county_split = county.split(" County")
    california_data.loc[index, "County"] = str(county_split[0])
        
california_data.head()
```

```python
population_2011_df = california_data[["County","Population"]]
population_2011_df = population_2011_df.reset_index(drop=True)
population_2011_df.head()
len(population_2011_df)
```

```python
fips_df = pd.read_csv('fips_modified.csv', dtype={'Fips': object})
fips_df = fips_df[["County","Fips"]]
len(fips_df)
```

```python
merge_df = pd.merge(population_2011_df, fips_df, on="County")
merge_df
```

```python
# Save as a csv
# Note to avoid any issues later, use encoding="utf-8"
merge_df.to_csv("population_2011.csv", encoding="utf-8", index=False, header=True)
```
